```python
#import scipy.optimize
import numpy as np
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

```

```python
class CustomModelClassifier:
  def __init__(self, rfc, svclass):
    self.rfc = rfc
    self.SVClass  = svclass 
    return

  def train(self,X_train,y_train):
    self.rfc.fit(X_train,y_train)
    X_svc_train = self.fetchIndVectors(X_train)
    print(X_svc_train)
    nsamples, nx, ny = X_svc_train.shape
    X_svc_train =  X_svc_train.reshape((nsamples,nx*ny))
    self.SVClass.fit(X_svc_train,y_train)
    return

  def test(self,x_test):
    fi = self.fetchIndVectors(x_test)
    return self.SVClass.predict(fi)

  def fetchIndVectors(self,x_train):
    trees = self.rfc.estimators_
    yProb = list()
    for tree in trees:
        yProb.append([self.binarize(x) for x in tree.predict_proba(x_train)])
    yProb = np.array(yProb)    
    return yProb

  def binarize(self,vec):
    maxVal = max(vec)
    flag = 0
    for idx in range(len(vec)):
      val = vec[idx]
      if val < maxVal and val != 0:
        vec[idx] = 0
      elif flag == 0:
        vec[idx] = 1
        flag = 1
      else:
        vec[idx] = 0
    return vec





'''
def hinge_loss(t,y):
  #def max(0,1-t*y)
  return [max(0,1-t_i*y_i) for t_i,y_i in zip(t,y)]

def error_f(W,c_n,t_mat,y_mat):
  
  return (1/2) * (sqrt(np.matmul(np.transpose(W),W))) + c_n* sum(hinge_loss(t_mat,y_mat))
''' 
```
